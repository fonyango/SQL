#### 1. How many distinct players are in this table?

```python
%%sql

SELECT COUNT(*) FROM soccer;
```

#### 2. How many distinct nationalities are there?

```python
%%sql

SELECT COUNT (DISTINCT nationality) 
AS 'Number of Nationalities' 
FROM soccer;
```

#### 3. How many players come from each nationality?


Top 10 nationalities based on the number of palyers

```python
%%sql

SELECT nationality, COUNT(nationality)
FROM soccer
GROUP BY nationality
ORDER BY COUNT(nationality)
DESC
LIMIT 10;
```

The last 10 nationalities based on the number of players

```python
%%sql

SELECT nationality, COUNT(nationality)
FROM soccer
GROUP BY nationality
ORDER BY COUNT(nationality)
ASC
LIMIT 10;
```

#### 4. How many distinct club_teams are there? And How many players come form each club?

Number of club teams

```python
%%sql

SELECT COUNT (DISTINCT club_team) 
AS 'Number of Club Teams' 
FROM soccer;
```

Top ten club teams based on the number of players

```python
%%sql

SELECT club_team, COUNT(club_team)
FROM soccer
GROUP BY club_team
ORDER BY COUNT(club_team)
DESC
LIMIT 10;
```

Bottom 10 club teams based on the number of players

```python
%%sql

SELECT club_team, COUNT(club_team)
FROM soccer
GROUP BY club_team
ORDER BY COUNT(club_team)
ASC
LIMIT 10;
```

#### 5. Find out the rank of club_teams based on the average overall_rating of players whose overall_rating is greater than 80

```python
%%sql

SELECT club_team, AVG(overall_rating)
FROM soccer
WHERE overall_rating > 80
GROUP BY club_team
ORDER BY AVG(overall_rating)
DESC;
```

#### 6. Find out overall_rating, wage and value of players in Manchester United ranked based on their overall_rating

```python
%%sql

SELECT name, overall_rating, wage_euro, value_euro
FROM soccer
WHERE club_team = 'Manchester United'
ORDER BY overall_rating DESC, value_euro;
```

#### 7. Find out overall_rating, wage and value of players in Arsenal ranked based on their overall_rating

```python
%%sql

SELECT name, overall_rating, wage_euro, value_euro
FROM soccer
WHERE club_team = 'Arsenal'
ORDER BY overall_rating DESC, value_euro;
```

#### 8. What are the club_team, overall_rating, club_position and dribling score of players whose prefered foot is left

```python
%%sql

SELECT name, club_team, overall_rating, club_position, dribbling
FROM soccer
WHERE preferred_foot = 'Left'
ORDER BY overall_rating DESC
LIMIT 25;
```
